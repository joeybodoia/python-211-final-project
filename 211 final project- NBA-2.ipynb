## Plot all metrics as a function of each unique season and look for trends in NBA over time

```python
#create a new column for the data frame that is total shots made
```

```python
#create a new column for the data frame that is total shots attempted
```

```python
nba_df['Total Shots Made'] = nba_df.apply(lambda row: row['FGM'] + row['3PM'], axis=1)
nba_df.head()
```

```python
nba_df['Total Shots Attempted'] = nba_df.apply(lambda row: row['FGA'] + row['3PA'], axis=1)
nba_df.head()
```

```python
#group the nba df by season and take a league-wide average of each metric for each season year 
```

```python
by_season = nba_df.groupby('season')
average_stats_df = by_season.agg(
    {
        'WIN%': ['mean'],
        'Minutes': ['mean'],
        'Points': ['mean'],
        'FGM': ['mean'],
        'FGA': ['mean'],
        'FG%': ['mean'],
        '3PM': ['mean'],
        '3PA': ['mean'],
        '3P%': ['mean'],
        'FTM': ['mean'],
        'FTA': ['mean'],
        'FT%': ['mean'],
        'OREB': ['mean'],
        'DREB': ['mean'],
        'REB': ['mean'],
        'AST': ['mean'],
        'TOV': ['mean'],
        'STL': ['mean'],
        'BLK': ['mean'],
        'BLKA': ['mean'],
        'PF': ['mean'],
        'PFD': ['mean'],
        '+/-': ['mean']
        
    }
)
average_stats_df.head()
```

```python
#by_season = nba_df.groupby('season')
#by_season.mean().head()
```

```python
by_season.mean().reset_index().plot(x='season', y='Points')
by_season.mean().reset_index().plot(x='season', y='Total Shots Made')
by_season.mean().reset_index().plot(x='season', y='Total Shots Attempted')
```

```python
#Strong upward trend in points because NBA teams are taking many more shots per game
```

```python
by_season.mean().reset_index().plot(x='season', y='FGM')
by_season.mean().reset_index().plot(x='season', y='3PM')
by_season.mean().reset_index().plot(x='season', y='FGA')
by_season.mean().reset_index().plot(x='season', y='3PA')
```

```python
#conclusion: NBA players are shooting a lot more shots, specifically more 3 point shots... shift in the way the game of basketball is being played in the NBA, more emphasis on 3 point shots 
```

```python
by_season.mean().reset_index().plot(x='season', y='FG%')
by_season.mean().reset_index().plot(x='season', y='3P%')
```

```python
by_season.mean().reset_index().plot(x='season', y='PF')
by_season.mean().reset_index().plot(x='season', y='FTM')
by_season.mean().reset_index().plot(x='season', y='FTA')
by_season.mean().reset_index().plot(x='season', y='FT%')
```

```python
#conclusion: Refs might not be calling the games as strictly; allowing more contact. 
```

```python
by_season.mean().reset_index().plot(x='season', y='REB')
by_season.mean().reset_index().plot(x='season', y='OREB')
by_season.mean().reset_index().plot(x='season', y='DREB')
```

```python
#Interesting how Rebounds has significantly increased over time, but Offensive Rebs has gone down while efensive Rebs has gone up. Indicates that NBA players are much less inclined to go for offensive rebounds in today's game as opposed to late 90s and early 2000s 
```

```python
by_season.mean().reset_index().plot(x='season', y='+/-')
by_season.mean().reset_index().plot(x='season', y='AST')
by_season.mean().reset_index().plot(x='season', y='TOV')
by_season.mean().reset_index().plot(x='season', y='STL')
```

```python
#interesting that assists has gone up in the last 8 years but turnovers has gone down. Downtrend in turnovers started before increase in assits 
```

```python
#What has lead to the decrease in turnovers? Better coaching? Defense could be getting lazier
```
